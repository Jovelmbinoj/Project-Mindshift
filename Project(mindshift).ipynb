```python
#installing libraries
import tensorflow as tf
import numpy as np
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
```

```python
# Load and preprocess data
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
#reshape
x_train = np.expand_dims(x_train, -1)
x_test = np.expand_dims(x_test, -1)
```

```python
# Convert to RGB & resize
x_train = tf.image.resize(tf.image.grayscale_to_rgb(tf.convert_to_tensor(x_train)), (32, 32))
x_test = tf.image.resize(tf.image.grayscale_to_rgb(tf.convert_to_tensor(x_test)), (32, 32))
```

```python
#Normalizing
x_train = x_train / 255
x_test = x_test / 255
```

```python
# Encoding
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
#defining model
model_1 = VGG16(weights='imagenet', include_top=False, input_shape=(32, 32, 3))
model_1.trainable = False
x = Flatten()(model_1.output)
x = Dense(1024, activation='relu')(x)
x = Dropout(0.5)(x)
x = Dense(512, activation='relu')(x)
x = Dropout(0.5)(x)
x = Dense(10, activation='softmax')(x)
model = Model(inputs=model_1.input, outputs=x)
```

```python
# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
model.fit(x_train, y_train, epochs=10, batch_size=32)
```
